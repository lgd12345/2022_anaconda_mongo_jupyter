```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from wordcloud import STOPWORDS, WordCloud
import nltk
```

```python
# nltk.download()
```

```python
# riss_bigdata.csv에서 제목 읽어서 출력
all_data = pd.read_csv('riss_bigdata.csv',encoding='utf-8')
all_data
```

```python
all_title = all_data['제목']
all_title
```

```python
# stopwords.words('english')
# 중복제거 (set 중복허용안함)
```

```python
stopWords = set(stopwords.words('english'))
stopWords
```

```python
first_title = all_title[0]
first_title
```

```python
# 영문자가 아닌 것 제거
EnWords = re.sub(r'[^a-zA-Z]+',' ',str(first_title))
EnWords
```

```python
# 토큰화 시키기
EnWordsToken = word_tokenize(EnWords.lower())
EnWordsToken
```

```python
# 명사가 아닌 것 빼기
EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
EnWordsTokenStop
```

```python
lemma = WordNetLemmatizer()
EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
EnWordsTokenStopLemma
```

```python
lemma = WordNetLemmatizer() #표제어 추출 객체
words = []

for title in all_title:
    EnWords = re.sub(r'[^a-zA-Z]+',' ',str(title))#영어가 아닌 것을 ' '으로 대체 (^ : not)
    EnWordsToken = word_tokenize(EnWords.lower()) #소문자로
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop] #기본형태
    words.append(EnWordsTokenStopLemma)
words
```

```python
# 전처리가 끝난 words 를 1차원 list
# 람다로 표현할 것임
from functools import reduce
words2 = list(reduce(lambda x,y : x+y, words))
words2
```

```python
# 단어 빈도수 구하기
from collections import Counter
count = Counter(words2)
count
```

```python
# 상위 50개 단어 딕셔너리에 저장
word_count = dict()
# count.most_common(50)
```

```python
# 50개 중에서 글자가 5 글자보다 많은 것
for tag, counts in count.most_common(50):
    if(len(str(tag))) > 5:
        word_count[tag] = counts
word_count
```

```python
import matplotlib.pyplot as plt
# 히스토그램 그리기
plt.figure(figsize=(10,5))
sorted_Keys = sorted(word_count, key=word_count.get,reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)
plt.bar(range(len(word_count)),sorted_Values,align='center')
plt.xticks(range(len(word_count)),list(sorted_Keys),rotation='90')
plt.show()
```
