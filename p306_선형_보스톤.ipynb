```python
# 보스톤 주택가격 데이터에 머신러닝 기반의 회귀 분석 수행(p306)

from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()
boston.DESCR
```

```python
boston.data
```

```python
# 컬럼네임
boston.feature_names
```

```python
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df
```

```python
# CRIM: 지역별 범죄 발생률
# ZN: 25000평방 피트를 초과하는 거주지역
# INDUS: 비상업지역의 넓이 비율
# CHAS: 찰스강의 더미 변수(1은 강의 경계 0은 경게 아님)
# NOX: 일산화질소 농도
# RM :거주할 수 있는 방 개수
# AGE: 1940 이전에 건축된 주택 비율
# DIS: 5개 주요 고용센터까지 가중 거리
# TAX: 10000 달러당 재산세 비율
# PTRATIO: 지역의 교사와 학생수 비율
# B: 지역의 흑인 거주비율
# LSTAT: 하위 계층의 비율
```

```python
# boston_df 에서 PRICE 칼럼 추가 (boston.target)
boston_df['PRICE'] = boston.target
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# x,y (data, label)분할하기
Y = boston_df['PRICE']
X = boston_df.drop(['PRICE'],axis = 1, inplace=False) #bost axis 1은 열, 0은 행 / inplace : 대체 할 것
```

```python
#훈련용 데이터와 평가용 데이터 분할하기
# 테스트용 비율 0.3 : 30퍼
# random_state=156 랜덤데이터를 고정시킨다.
X_train, X_test,Y_train, Y_test = train_test_split(X,Y,test_size=0.3,
                                                   random_state=156)
```

```python
#선형 회귀 분석 모델
lr = LinearRegression()
# 훈련 X_train,Y_train 문제와 답 알려줌
lr.fit(X_train,Y_train)
```

```python
# 선형 회귀 분석 : 평가 데이터에 대한 예측 수행 ==> 예측 결과
Y_predict = lr.predict(X_test)
```

```python
Y_predict
# 예측한 답
```

```python
# 예측한 답과 실제답을 비교해 봐야한다. 차이가 클수록 않좋음
# 오차의 제곲값으로 처리하는 import mean_squared_error(사이킷런 라이브러리) : MSE(평가지표)
#                                             r2_scor(사이킷런 라이브러리) : R2(결정계수)
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
# 실제값 : Y_test 예측값 : Y_predict
mse = mean_squared_error(Y_test,Y_predict)
rmse = np.sqrt(mse)
```

```python
#결정 게수 (0 <= R2 <= 1) : 1에 가까울 수록 회귀식정확도가 높다로 해석
print('mse: ',mse,'rmse: ',rmse)
print('R^2 : ',r2_score(Y_test,Y_predict))
```

```python
print('Y절편 값 : ', lr.intercept_)
print('회귀계수(기울기)',np.round(lr.coef_,1))
```

```python
coef = pd.Series(data=np.round(lr.coef_,2),index = X.columns)
coef.sort_values(ascending=False)
```
