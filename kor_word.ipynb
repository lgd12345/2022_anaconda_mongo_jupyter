```python
import json
import re
from konlpy.tag import Okt
from collections import Counter
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from wordcloud import WordCloud
from docutils.parsers.rst.directives import encoding
```

```python
data = json.loads(open('4차 산업혁명.json','r',encoding='utf-8').read())
data
```

```python
# 0번째 message의 내용 뽑기
data[0]['message']
```

```python
# 모든 message의 내용 뽑기
for item in data:
    print(item['message'])
```

```python
# 정규식을 이용해 (문자나 숫자) : (\w)가 아닌 것은 ' ' 공백으로 
message =''
for item in data:
    if 'message' in item.keys():
        message = message + re.sub(r'[^\w]',' ', item['message'])
message
```

```python
from konlpy.tag import Okt
```

```python
# 명사만 가져오기
nlp = Okt()
message_N = nlp.nouns(message)
message_N
```

```python
# 단어당 개수 구하기
count = Counter(message_N)
count
# 글자수가 2개 이상인 것 중에 80개 추출하여 막대그래프, 워드클라우드
```

```python
word_count = dict()
# 80개 중에서 글자가 5 글자보다 많은 것
for tag, counts in count.most_common(80):
    if(len(str(tag))) >= 2:
        word_count[tag] = counts
word_count
```

```python
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
# 폰트 가져오고
font = font_manager.FontProperties(fname=font_path).get_name()
# 폰트 적용
# matplotlib.rc('font', family=font) 이렇게 적어도 됨
rc('font', family=font)

# 히스토그램 그리기
plt.figure(figsize=(12,5))
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid(True) #세로 선 생김
sorted_Keys = sorted(word_count, key=word_count.get,reverse=True) # reverse=True 이거 안 적으면 오름차순
sorted_Values = sorted(word_count.values(), reverse=True)
plt.bar(range(len(word_count)),sorted_Values,align='center')
plt.xticks(range(len(word_count)),list(sorted_Keys),rotation='90')
plt.show()
```

```python
from wordcloud import STOPWORDS, WordCloud
stopwords = set(STOPWORDS)

wc = WordCloud(font_path = "C:/Windows/Fonts/malgun.ttf",background_color='ivory',stopwords=stopwords, width=1000,height=800)
cloud = wc.generate_from_frequencies(word_count)
# cloud
f = plt.figure(figsize=(8,8))
plt.imshow(cloud)
# 테두리 안 하기
plt.axis('off')
plt.show()
# f.savefig('kor_word.png')
```
